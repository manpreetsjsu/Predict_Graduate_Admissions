```python
#Reference
# https://www.kaggle.com/kralmachine/analyzing-the-graduate-admission-eda-ml
import os
import pandas as pd
import numpy as np

#graph plot libraries
import matplotlib.pyplot as plt
import seaborn as sns
#ML
from sklearn.model_selection import train_test_split
import statsmodels.formula.api as smf
import statsmodels.api as sm

from sklearn.metrics import confusion_matrix, classification_report, precision_score, accuracy_score
from sklearn.neighbors  import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix

os.getcwd()
```

```python
df1 = pd.read_csv("Admission_Predict.csv")
print(df1.shape)
df2 = pd.read_csv("Admission_Predict_Ver1.1.csv")
print(df2.shape)
frames = [df1,df2]
df = pd.concat(frames,sort=False)
df.shape
```

```python
#split train data, test data
df_X = df # training data
df_Y = df['ChanceOfAdmit']  #actual Y values for train data 
```

```python
X_train,X_test,y_train,y_test=train_test_split(df_X,df_Y,test_size=0.2,random_state=42)
print('X_train',X_train.shape)
print('X_test',X_test.shape)
print('y_train',y_train.shape)
print('y_test',y_test.shape)
```

```python
#applying logisitic regression
#Now that we have split our data into appropriate sets, let's write down the formula to be used for the logistic regression.
# Create a string for the formula
cols = X_train.columns.drop('ChanceOfAdmit')
formula = "ChanceOfAdmit ~ " + ' + '.join(cols)
print(formula, '\n')
X_train.columns
```

```python
# Run the model and report the results

model = smf.glm(formula= formula,data=X_train, family=sm.families.Binomial())
logistic_fit = model.fit()

print(logistic_fit.summary())
```

```python
# predict the test data and show the first 5 predictions
predictions = logistic_fit.predict(X_test)
predictions[1:10]
```

```python
# Convert these probabilities into nominal values and check the first 5 predictions again.
predictions_nominal = [ 0 if x < 0.5 else 1 for x in predictions]
predictions_nominal[1:10]
```

```python
#convert y_test values into binary values because our training data dependent variable is continuous
# we are doing classification using logistic regression
y_test_binary = [0 if x<0.5 else 1 for x in y_test]
y_test.shape
```

```python
#verify prediction results with test observation results


print(classification_report(y_test_binary, predictions_nominal, digits=3))

cfm = confusion_matrix(y_test_binary, predictions_nominal)

true_negative = cfm[0][0]
false_positive = cfm[0][1]
false_negative = cfm[1][0]
true_positive = cfm[1][1]

print('Confusion Matrix: \n', cfm, '\n')

print('True Negative:', true_negative)
print('False Positive:', false_positive)
print('False Negative:', false_negative)
print('True Positive:', true_positive)
print('Correct Predictions', 
      round((true_negative + true_positive) / len(predictions_nominal) * 100, 1), '%')
print("Accuracy of model is ", accuracy_score(y_test_binary,predictions_nominal))
```
